```python
import numpy as np
import matplotlib.pyplot as plt

# Вхідні дані
x = np.array([0.1, 0.3, 0.4, 0.6, 0.7])
y = np.array([3.2, 3.0, 1.0, 1.8, 1.9])

# Побудова полінома 4-го степеня
coeffs = np.polyfit(x, y, 4)
poly = np.poly1d(coeffs)

print("Інтерполяційний поліном:")
print(poly)

# Побудова графіка
x_new = np.linspace(0.1, 0.7, 100)
y_new = poly(x_new)

plt.scatter(x, y, label="Задані точки")
plt.plot(x_new, y_new, label="Поліном")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.show()

# Обчислення значень
print("y(0.2) =", poly(0.2))
print("y(0.5) =", poly(0.5))
```

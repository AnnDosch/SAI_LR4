```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Вхідні дані (заміни на свій варіант)
X = np.array([0, 5, 10, 15, 20, 25]).reshape(-1, 1)
Y = np.array([21, 39, 51, 63, 70, 90])

# Навчання моделі
model = LinearRegression()
model.fit(X, Y)

# Коефіцієнти регресії
b0 = model.intercept_   # вільний член
b1 = model.coef_[0]     # коефіцієнт при x

print("Рівняння регресії:")
print(f"y = {b0:.3f} + {b1:.3f}x")

# Прогноз значень
Y_pred = model.predict(X)

# Графік
plt.scatter(X, Y, label="Експериментальні точки")
plt.plot(X, Y_pred, label="Лінія регресії")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
